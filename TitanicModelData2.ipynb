```python
import pandas as pd
train_data=pd.read_csv("train.csv")
test_data=pd.read_csv("test.csv")
```

```python
sex=pd.get_dummies(train_data['Sex'], drop_first=True)
embarked=pd.get_dummies(train_data['Embarked'], drop_first=True)
pclass=pd.get_dummies(train_data['Pclass'], drop_first=True)

train_data = pd.concat([train_data,sex,embarked,pclass],axis=1)
```

```python
train_data.drop(['Sex', 'Embarked','Pclass'], axis=1, inplace=True)
```

```python
train_df = train_data.drop(['Name'], inplace=True, axis=1)
```

```python
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].mean())
```

```python
test_data.isnull().sum()
```

```python
train_data.drop(['Cabin','Ticket'], axis=1, inplace=True)
```

```python
x_train=train_data.drop('Survived', axis=1)
y_train=train_data['Survived']
```

```python
test_data.head()
```

```python
sex1=pd.get_dummies(test_data['Sex'], drop_first=True)
embarked1=pd.get_dummies(test_data['Embarked'], drop_first=True)
pclass1=pd.get_dummies(test_data['Pclass'], drop_first=True)
```

```python
test_data = pd.concat([test_data,sex1,embarked1,pclass1],axis=1)
```

```python
test_data.drop(['Sex', 'Embarked','Pclass','Name','Cabin','Ticket'], axis=1, inplace=True)
```

```python
train_data.drop(['Fare'], axis=1, inplace=True)
test_data.drop(['Ticket'], axis=1, inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model=LogisticRegression()
```

```python
lr_model.fit(x_train,y_train)
```

```python
test_data['Age'] = test_data['Age'].fillna(test_data['Age'].mean())
```

```python
test_data.dropna(inplace=True)
```

```python
yhat=lr_model.predict(test_data)
```

```python
print(type(yhat))
```

```python
yhat_df=pd.DataFrame(yhat)
```
